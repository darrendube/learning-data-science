```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
# "hue" separates the points by the selected hue - species in this case
# SCATTERPLOT - x=numerical, y=numerical
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
```

```python
# lineplot only requires the data as a parameter. ['species'] column is dropped because we don't need it
# LINEPLOT - x=categorical/ordinal, y=numerical
sns.lineplot(data=iris.drop(["species"], axis=1))
```

```python
# KDEPLOT - x=numerical, y=numerical
sns.kdeplot(data=iris.drop(['species'], axis=1))
```

```python
# simple barchart. Shows how many times a particular item occurs in the data
# BARCHART - x=categorical, y=numerical
sns.countplot(x=iris['species'])
```

```python
# BOX-PLOT - x=categorical, y=numerical
sns.boxplot(x='species', y='sepal_length', data=iris)
```

```python
# shows distribution better than boxplot
# VIOLINPLOT - x=categorical, y=numerical
sns.violinplot(x='species', y='sepal_length', data=iris)
```

```python
# represents actual points instead of distributions.
# CATEGORICAL SCATTERPLOT - x=categorical, y=numerical
sns.catplot(x='species', y='sepal_length', data=iris)
```

```python
# jitter=False removes the horizontal separation of points that lie on the same position.
# disadvantage - many points are hidden because some are on the same position
sns.catplot(x='species', y='sepal_length', data=iris, jitter=False)
```

```python
# shows mean of each category, unlike countplot which shows frequency of each category
# line above bar is the error/uncertainty in calculation. Confidence interval or standard deviation
sns.barplot(x='species', y='sepal_length', data=iris)
```

```python
# correlation matrix for iris
iris.corr()
```

```python
# can be visual representation of correlation matrix, or any other table with same layout
# HEATMAP - takes DataFrame or any 2D dataset. In case of DataFrame, row and column labels are used
sns.heatmap(iris.corr(), annot=True)
```

```python
# petal_length and sepal_length have 0.87 correlation. Below is a scatterplot of these two variables
sns.scatterplot(x='petal_length', y='sepal_length', data=iris)
```

```python
# petal_width and sepal_width have -0.37 correlation
sns.scatterplot(x='petal_width', y='sepal_width', data=iris)
```

```python
# combines multiple different subplots in same figure
sns.FacetGrid(iris, col="species").map(sns.kdeplot, 'petal_length')
```

```python
# like a correlation matrix, but scatter plot of the two variables is plotted instead
# in the diagonal, a distribution plot is drawn (therefore there is one distribution plot per variable)
sns.pairplot(iris)
```
